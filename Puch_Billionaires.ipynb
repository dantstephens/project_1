```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import numpy as np
import hvplot.pandas
import pandas as pd
```

```python
# Load the CSV file into a Pandas DataFrame
billion = pd.read_csv("../project_1/Final_Billionaires_Statistics.csv")
billion_df = pd.DataFrame(billion)

#convert final worth to billions
billion_df['finalWorth'] = 1000000 * billion_df['finalWorth']

# Display sample data
billion_df.head()
```

```python
#Male billionaire DataFrame:
male_billion = billion_df['gender'] == 'M'
male_billion_df = billion_df[male_billion]
male_billion_df.head()
```

```python
#Female billionaire DataFrame:
female_billion = billion_df['gender'] == 'F'
female_billion_df = billion_df[female_billion]
female_billion_df.head()
```

```python
#Male Vs Female totals
male_count = male_billion_df['gender'].count()
female_count = female_billion_df['gender'].count()
pie_labels = ['Male', 'Female']
pie_chart = [male_count, female_count]
colors = ["Lightskyblue", "lightcoral"]

#Male VS Female Pie Chart 
plt.pie(pie_chart, labels=pie_labels, colors=colors, autopct="%1.1f%%", startangle=140)
plt.axis("equal")
plt.title('Male to Female Billionaires')
plt.show()
```

```python
pd.options.display.float_format= '{:.0f}'.format
#Male Vs Female Median Worth Totals
male_worth_median = male_billion_df['finalWorth'].median()
female_worth_median = female_billion_df['finalWorth'].median()

#Male Vs Female Mode Worth Totals
male_worth_mode = male_billion_df['finalWorth'].mode().values[0]
female_worth_mode = female_billion_df['finalWorth'].mode().values[0]

#Male Vs Female Count Total
male_count = male_billion_df['finalWorth'].count()
female_count = female_billion_df['finalWorth'].count()

#Male Vs Female Mean Total
male_mean = male_billion_df['finalWorth'].mean()
female_mean = female_billion_df['finalWorth'].mean()

#Male Vs Female Avg Age
male_age_mean = male_billion_df['age'].mean()
female_age_mean = female_billion_df['age'].mean()

#Male Vs Female Max
male_worth_max = male_billion_df['finalWorth'].max()
female_worth_max = female_billion_df['finalWorth'].max()

#Male vs Female Min
male_worth_min = male_billion_df['finalWorth'].min()
female_worth_min = female_billion_df['finalWorth'].min()

worth_data = {
    'Gender': ['Male', 'Female'],
    'Count': [male_count, female_count],
    'Average Age' : [male_age_mean, female_age_mean],
    'Median Worth': [male_worth_median, female_worth_median],
    'Mode Worth': [male_worth_mode, female_worth_mode],
    'Mean Worth' : [male_mean , female_mean],
    'Max Worth' : [male_worth_max, female_worth_max],
    'Min Worth' : [male_worth_min, female_worth_min]
}

worth_summary_df = pd.DataFrame(worth_data)
worth_summary_df
```

```python
#Billionaires Male vs Female by industry data
#Male industry count
industries_male = male_billion_df["industries"].value_counts()
#Female industry count
industries_female = female_billion_df['industries'].value_counts()
#full dataframe for male vs female by industry
compare_industries_df = pd.DataFrame({'Male Billionaires': industries_male, 'Female Billionaires': industries_female})
#fill NaN values with 0
compare_industries_df = compare_industries_df.fillna(0)
#reset the index
compare_industries_df.reset_index(inplace=True)
#show the data
compare_industries_df
```

```python
#Bar graph comparing the Billionaires Male vs Female by Industry 
compare_industries_bar = compare_industries_df.plot(kind= 'bar', x='industries', color=['Lightskyblue', 'lightcoral'], figsize=(12,6), title= 'Billionaires Male vs Female by Industry', xlabel= 'Industries')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
#Billionaires Male vs Female by Country data
#Male industry count
country_male = male_billion_df["country"].value_counts()
#Female industry count
country_female = female_billion_df['country'].value_counts()
#full dataframe for male vs female by industry
compare_country_df = pd.DataFrame({'Male Billionaires': country_male, 'Female Billionaires': country_female})
#fill NaN values with 0
compare_country_df = compare_country_df.fillna(0)
#reset the index
compare_country_df.reset_index(inplace=True)
#show the data
compare_country_df
```

```python
#Notes
compare_country_df.sort_values('Male Billionaires')
```

```python
#Male billionaires globally
#Merge country count data with origional data to get the lat and lon of the country
location_merge = pd.merge(compare_country_df, billion_df, on='country', how='left')
#create new dataframe with only columns we will use for the map
location_fields = location_merge[['Male Billionaires','country','longitude_country', 'latitude_country']]
coordinates_df = pd.DataFrame(location_fields)
#keep only the unique countries
uniq_coordinates_df = coordinates_df.drop_duplicates(subset='country')
clean_male_uniq_coordinates_df = uniq_coordinates_df [uniq_coordinates_df['Male Billionaires'] != 0]


map_plot_male = clean_male_uniq_coordinates_df.hvplot.points(
    "longitude_country",
    "latitude_country",
    geo = True,
    tiles = "OSM",
    frame_width = 700,
    frame_height = 500,
  
)
map_plot_male
```

```python
clean_male_uniq_coordinates_df.head()
```

```python
#Female billionaires globally 
#Merge country count data with origional data to get the lat and lon of the country
female_location_merge = pd.merge(compare_country_df, billion_df, on='country', how='left')
#create new dataframe with only columns we will use for the map
female_location_fields = female_location_merge[['Female Billionaires','country','longitude_country', 'latitude_country']]
female_coordinates_df = pd.DataFrame(female_location_fields)
#keep only the unique countries
female_uniq_coordinates_df = female_coordinates_df.drop_duplicates(subset='country')
clean_female_uniq_coordinates_df = female_uniq_coordinates_df [female_uniq_coordinates_df['Female Billionaires'] != 0]


map_plot_female = clean_female_uniq_coordinates_df.hvplot.points(
    "longitude_country",
    "latitude_country",
    geo = True,
    tiles = "OSM",
    frame_width = 700,
    frame_height = 500
  
)
map_plot_female
```

```python
condition_usa = male_billion_df['country'] == 'United States'
state_male = pd.DataFrame(male_billion_df[condition_usa])
state_male_bar = state_male[['state']]
sorted_male_bar = state_male_bar.sort_values(by='state')
state_male_bar_count = sorted_male_bar.value_counts()
top_20_state_male = state_male_bar_count.head(20)
top_20_state_male_df = pd.DataFrame(top_20_state_male)
top_20_state_male_df.reset_index(inplace=True)
top_20_state_male_df.head()

top_20_state_male_df.plot(kind = 'bar', x='state', color= ['lightskyblue'], figsize= (12, 6), title='Male Billionaires Top 20 States', xlabel='State')
plt.xticks(rotation=45, ha='right')
plt.show
```

```python
condition_usa = female_billion_df['country'] == 'United States'
state_female = pd.DataFrame(female_billion_df[condition_usa])
state_female_bar = state_female[['state']]
sorted_female_bar = state_female_bar.sort_values(by='state')
sorted_female_bar.count()
state_female_bar_count = sorted_female_bar.value_counts()
top_20_state_female = state_female_bar_count.head(20)
top_20_state_female_df = pd.DataFrame(top_20_state_female)
top_20_state_female_df.reset_index(inplace=True)
top_20_state_female_df.head()

top_20_state_female_df.plot(kind = 'bar', x='state', color= ['lightcoral'], figsize= (12, 6), title='Female Billionaires Top 20 States', xlabel='State')
plt.xticks(rotation=45, ha='right')
plt.show
```
