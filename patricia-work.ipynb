```python
# Dependencies
import pandas as pd
from pathlib import Path
#charts
import matplotlib.pyplot as plt
import numpy as np
```

```python
# import
billionaires_file = Path("Final_Billionaires_Statistics.csv")
```

```python
# read
billionaires_file_df = pd.read_csv(billionaires_file)
billionaires_file_df.head()
```

```python
billionaires_file_df.describe()
```

```python
#showing the columns I will be focusing on
billionaires_file_df[["rank","personName","selfMade","industries","country","finalWorth"]].head(10)
```

```python
# Counting self made vs. unearned/inherited wealth
count = billionaires_file_df["selfMade"].value_counts()
count
```

```python
#pie chart
labels = ["True", "False"]
sizes = [1812, 828]
plt.figure(figsize=(8,6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Selfmade vs. Inherited')
plt.show()
```

```python
billionaires_file_df.groupby(['industries','selfMade']).count()["rank"]
```

```python
Categories = ['Finance & Investments', 'Manufacturing', 'Technology']
Counts = [276, 235, 292]
plt.figure(figsize=(8, 6))
plt.bar(Categories, Counts)
plt.xlabel('SelfMade')
plt.ylabel('# of Billionaires')
plt.title('SelfMade Billionaires')
plt.show()
```

```python
Categories = ['Finance & Investments', 'Manufacturing', 'Technology']
Counts = [96, 89, 22]
plt.figure(figsize=(8, 6))
plt.bar(Categories, Counts)
plt.xlabel('Inherited')
plt.ylabel('# of Billionaires')
plt.title('Inherited Billionaires')
plt.show()
```

```python
# Counting countries self made vs inherited
billionaires_file_df.groupby(['country','selfMade']).count()["rank"]
```

```python
Categories = ['True', 'False']
Counts = [540, 214]
plt.figure(figsize=(8, 6))
plt.bar(Categories, Counts)
plt.xlabel('SelfMade vs. Inherited')
plt.ylabel('# of Billionaires')
plt.title('SelfMade vs. Inherited Billionaires in the U.S')
plt.show()
```
